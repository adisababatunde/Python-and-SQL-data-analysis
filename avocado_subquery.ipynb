```python
#load in data
import pandas as pd
df= pd.read_csv('avocado.csv')
df.head()
```

```python
# 2 times Average Price for 2018 --- extremely high price
yr2018 = df['year']==2018
avg_2times_2018 = df[yr2018]['AveragePrice'].mean()*2.0
```

```python
#filtering the df
condition = df['AveragePrice'] > avg_2times_2018
df[condition][['Date','AveragePrice']].sort_values(by='AveragePrice',ascending=False)
```
