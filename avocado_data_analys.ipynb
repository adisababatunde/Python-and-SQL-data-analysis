```python
#load in data
import pandas as pd
data = pd.read_csv('avocado.csv')
data.head()
```

```python
#select column
data.region
```

```python
#filtering data
filter = data[data['region'] == 'Orlando']
filter.head(5)
```

```python
#groupin data
data.groupby(['region'])['AveragePrice'].mean()
```

```python
data['AveragePrice'].mean()
```

```python
#conditional column
import numpy as np
data['price_category']=np.where(data['AveragePrice']>1.405,'expensive','inexpensive')
filter = data['price_category']== 'expensive'
data[filter].head()
```

```python
#select multiple columns
column_list = ['Date', 'Total Volume', 'region', 'price_category']
data[column_list].head()
```

```python
data.columns
```

```python
#unpivoting data
unpivot_table = data [['type','Small Bags','Large Bags']]
unpivot_table = pd.melt(unpivot_table,id_vars='type',value_vars=['Small Bags','Large Bags'])
unpivot_table.head(10)
```

```python
#Merging table
holidays = pd.read_csv('holiday.csv')
data = data.merge(holidays, on = 'Date', how ='left')
data
```

```python
#fill missing value in holiday
data['Holiday'] = data['Holiday'].fillna('normal')
```

```python
data.groupby('Holiday')['AveragePrice'].mean().sort_values(ascending=False)
```
